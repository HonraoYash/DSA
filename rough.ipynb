```python
from queue import Queue

class MyStack:
    def __init__(self):
        self.queue = Queue()

    def push(self, x: int) -> None:
        # We push the element in the queue, but since we want the 'pop' to work like a stack, all the elements 
        # in the queue before the newly added element are added to the end of the queue in the same order as they
        # appear.
        s = self.queue.qsize()
        self.queue.put(x)
        for i in range(s):
            self.queue.put(self.queue.get())

    def pop(self) -> int:
        return self.queue.get()

    def top(self) -> int:
        return self.queue.queue[0]

    def empty(self) -> bool:
        return not self.queue.full()

obj = MyStack()
obj.push(1)
obj.push(2)
obj.push(3)
print(obj.pop())
print(obj.top())
print(obj.empty())
```

```python
def uniformArray(nums1: list[int]) -> bool:
        nums2 = [1] * len(nums1)
        # Check for odd
        for i in range(len(nums1)):
            for j in range(0,i):
                if nums1[i] - nums1[j] >= 1:        
                    if (nums1[i] - nums1[j]) % 2 == 0:
                        continue
                    else:
                        nums2[i] = 0
                        break
            if nums2[i] == 0:
                continue
            if nums1[i] % 2 == 0:
                nums2[i] = 1
                break
            for j in range(i+1, len(nums1)):
                if nums1[i] - nums1[j] >= 1:        
                    if (nums1[i] - nums1[j]) % 2 == 0:
                        continue
                    else:
                        nums2[i] = 0
                        break

        if sum(nums2) == 0:
            return True

        nums2 = [0] * len(nums1)
        # Check for even
        for i in range(len(nums1)):
            for j in range(0,i):
                if nums1[i] - nums1[j] >= 1:
                    if (nums1[i] - nums1[j]) % 2 != 0:
                        continue
                    else:
                        nums2[i] = 1
                        break
            if nums2[i] == 1:
                continue
            if nums1[i] % 2 != 0:
                nums2[i] = 0
                break
            for j in range(i+1, len(nums1)):
                if nums1[i] - nums1[j] >= 1:        
                    if (nums1[i] - nums1[j]) % 2 != 0:
                        continue
                    else:
                        nums2[i] = 1
                        break
        if sum(nums2) == len(nums1):
            return True
        return False


nums1 = [2, 4, 6]
print(uniformArray(nums1))
```

```python
def can_all_be(parity, nums1):  # parity=1 for odd, parity=0 for even
        for i in range(len(nums1)):
            found = False
            # Option 1: use nums1[i] itself
            if nums1[i] % 2 == parity:
                found = True
            # Option 2: use nums1[i] - nums1[j] >= 1
            if not found:
                for j in range(len(nums1)):
                    if i == j:
                        continue
                    diff = nums1[i] - nums1[j]
                    if diff >= 1 and diff % 2 == parity:
                        found = True
                        break
            if not found:
                return False
            return True


nums1 = [2, 3, 8]
print( can_all_be(1, nums1) or can_all_be(0, nums1) )
```
