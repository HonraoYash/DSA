```python
from collections import Counter

nums = [1, 2, 3, 4, 5, 5, 5, 5, 4, 2, 1]
k = Counter(nums)

print(k.most_common(2))
```

```python
from collections import defaultdict

def maxActivated(points: list[list[int]]) -> int:

        r = len(points)
        adj = defaultdict(set)

        for i in range(r):
            if points[i][0] == points[i][1]:
                adj[points[i][0]].add(points[i])
            

        print(adj)

points = [[1,1],[1,2],[2,2]]
maxActivated(points)
```
